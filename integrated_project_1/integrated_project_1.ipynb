```python
# пишем функцию summary
def summary(df):
    print(df.info())
    display(df.head())
    count=0
    for col in df.columns:
        if df.isna().sum()[col]==0:
            count+=1
        else:
            print(f'Пропущенных значений в столбце {col}: {df.isna().sum()[col]}')
    if count==len(df.columns):
        print('Пропущенных значений в таблице нет')
    print(f'Повторов в таблице {df.duplicated().sum()}')  
    
summary(games)    
```

## Подготовка данных

### Замена названия столбцов

Произведём замену названия столбцов

```python
# производим замены в названии столбцов
games.columns = games.columns.str.lower()
```

### Обработка пропусков

В задании говорится о том, что нужно посмотреть на аббревиатуру `tbd` в столбце с оценкой пользователей. Посмотрим на уникальные значения столбца.

```python
# смотрим уникальные значения столбца
games['user_score'].unique()
```

`tbd` расшифровывается, как:To Be Determined, то есть "Будет определено". Значит оценка ещё не поставлена. Заменем эту аббревиатуру на Nan.

```python
# заменяем tbd на Nan
games['user_score'] = games['user_score'].replace('tbd', np.nan)
```

Посмотрим на пропуски в столбце `name`.

```python
# смротрим на пропуски в столбце 'name'
games.loc[games['name'].isna()]
```

Пропуски в столбце `name` и столбце `genre` на одних и тех же строках.

В столбце `year_of_release` 269 пропусков. Так как взять эти данные не представляется возможным, удалим эти строки и пропуски в столбце `name`. Всего удалим 271 строку из 16715, это 1,62% от всего датафрейма.

```python
# удаляем пропуски в столбцах 'name', 'year_of_release'
games = games.dropna(subset=['name', 'year_of_release']).reset_index(drop=True)
```

Пропуски в столбцах `critic_score`, `user_score`, `rating` не представляется возможным заменить. Так как их много и группировка по группам с заменой на медианные значения желаемого результата не даст. Можно попытаться сделать запрос в организацию ESRB. Но это не входит в учебную программу. Поэтому оставим эти пропуски. Будем работать с теми данными, что есть

### Замена типа данных

Изменим типы данных в столбцах: 
* `year_of_release` с вещественного не цельночисловое
* `user_score` с текстового в числовой
* `genre` из текстового в категориальный

```python
# переводим в целочисленное значение
games['year_of_release'] = games['year_of_release'].astype('Int64')
```

```python
# переводим из текстового в числовое значение
games['user_score']= games['user_score'].astype('float')
```

```python
# переводим из текстового в категориальное значение
games['genre']= games['genre'].astype('category')
```

### Изменения продаж по платформам с течением времени

Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построем распределение по годам(за весь период, 1980-2016). Определим за какой характерный срок появляются новые и исчезают старые платформы. Создадими сводную таблицу, в качестве индексов возьмём `platform` и определим топ платформ по продажам.

```python
# строим сводную таблицу и гистограмму распределения платформ
top_platform = games.pivot_table(index='platform', values='sum_sales', aggfunc='sum').sort_values(
by='sum_sales', ascending=False)

plt.figure(dpi=120)
plt.bar(x=top_platform.index, height=top_platform['sum_sales'])
plt.ylabel('Сумма продаж')
plt.xlabel('Имя платформы')
plt.xticks(rotation=90)
plt.title('Продажа платфор за всё время')
```

Отчетливо видны топ 6 платформ по продажам. Это:
* PS2
* X360
* Wii
* PS3
* DS 
* PS

Создадим список с топ 6 платформ.

```python
# создаем список с топ 6 платформ
top_6_platform = list(top_platform.reset_index().head(6)['platform'])
```

### Изменение продаж топ платформ за актуальный период

Посмотрим как изменялись продажи топ 6 платформ за период с 2005 года. Сделаем срез таблицы по двум критериям, год с 2005 и топ 6 платформ. Построим графики распределения.

```python
# исследуем продажи топ 6 платформ за период с 2005 года
games_actual = games.query('year_of_release >=2005')
games_actual_top_6 = games_actual[games_actual['platform'].isin(top_6_platform)]

plt.figure(figsize=(12, 7))
plt.grid(True)
plt.ylabel('Сумма продаж')
plt.xlabel('Год релиза')
plt.title('Продажа платфор топ 6 за всё время')

sns.lineplot(
data=games_actual_top_6,
x='year_of_release',
y='sum_sales',
hue='platform',
estimator='sum',
ci=None
);
```

**Вывод**

* исчезла платформа PS, значит после 2005 года игры для неё уже не выпускались
* заметно, что при выходе PS3 падают продажи  игр для PS2
* наибольшее число игр были проданы для платфор Sony PlayStation и Microsoft Xbox
* пик продаж приходится на 2008 - 2011 и наступает спустя 5 лет, после появления платформы
* на основе полученных данных предположим среднее время продолжительности жизни платформы: 6 - 8 лет.

### Как влияют на продажи внутри одной популярной платформы отзывы критиков

Из проведенного выше анализа ,в качестве популярноц платформы можно взять PS4 или X360. Построем диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

Хоть игр для PS4 продано больше, но число строк у X360 в 4 раза больше и доли пропущенных значений в оценках игр меньше.

В качестве популярной платформы, для исследования, возьмём платформу X360 от Microsoft. Напишем функцию, которая построит график диаграммы рассеивания продаж от оценок игр и посчитает значение корреляции.

```python
# пишем функцию для построения диаграммы рассеивания и расчета коэффициента корреляции
def user_critic_scatter(platform):
    games_platform = games_actual.query('platform == @platform')
    fig, ax = plt.subplots(1,2, figsize=(16, 6))
    fig.suptitle(f'Зависимость продаж от оценок критиков и пользователей для платформы {platform}', fontsize=20)
    
    ax1 = sns.scatterplot(
          data = games_platform,
        x='sum_sales',
        y='critic_score',
        ax=ax[0])
    ax1.set_title('Зависимость суммы продаж от оценки критиков')
    ax1.set_xlabel('Сумма продаж')
    ax1.set_ylabel('Оценки критиков')
    
    ax2 = sns.scatterplot(
          data = games_platform,
        x='sum_sales',
        y='user_score',
        ax=ax[1])
    ax2.set_title('Зависимость суммы продаж от оценки пользователей')
    ax2.set_xlabel('Сумма продаж')
    ax2.set_ylabel('Оценки пользователей')
    sns.set_style('darkgrid')
    
    print(f'Корреляция между продажами и оценкой критиков {platform}',
         round (games_platform['sum_sales'].corr(games_platform['critic_score']), 3))
    print(f'Корреляция между продажами и оценкой пользователей {platform}',
         round (games_platform['sum_sales'].corr(games_platform['user_score']), 3))
    
user_critic_scatter('X360')    
```

**Вывод**

* оценки пользователей никак не влияют на продажи, коэффициент корреляции равен 0.114
* оценки критиков слабо,но влияют на продажи, коэффициент корреляции 0.394, но это влияние заметно, при оценках от 60 и выше

## Проверка гипотез

Проверим 2 гипотезы:

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action  и Sports разные.

### Сравнение средних пользовательских рейтингов платформ

Нулевая гипотеза всегда формулируется так, чтобы использовать знак равенства. Другими словами, чтобы все данные лежали в районе изначально предполагаемого значения. Альтернативная гипотеза - это утверждение, которое принимается верным, если отбрасывается нулевая гипотеза. Если в альтерантивной гипотезе используестя значение не равно, то такая гипотеза является двухсторонней,потому что есть возможное отклонение и в большую, и в меньшую стороны.

Сформулируем нулевую гипотезу: **'Средние пользовательские рейтинги платформ Xbox One и PC равны'**.

Сформулируем альтернативную гипотезу: **'Средние пользовательские рейтинги платформ Xbox One и PC не равны'**.

Поскольку нужно оценить две выборки одной совокупности, применим метод о равенстве средних двух генеральных совокупностей. Применим метод scipy.stats.ttest_ind (array1, array2,equal_var), где:

array1, array2 — массивы, содержащие выборки; equal_var — параметр, задающий, считать ли равными дисперсии выборок.

Оценим размеры выборок, после удаления пропусков.

```python
# определяем размер выборок для платформы XOne
games_actual.query('platform == "XOne"')['user_score'].dropna().shape[0]
```

```python
# определяем размер выборок для платформы PC
games_actual.query('platform == "PC"')['user_score'].dropna().shape[0]
```

Выборки не равны, значит укажем параметр equal_var = False.

Применим t-test, для проверки нулевой гипотезы.

```python
alpha =.05   # критический уровень статистической значимости
#если p-value окажется меньше него - отвергнем гипотезу

tstat, pvalue = st.ttest_ind(games_actual.query('platform == "XOne"')['user_score'].dropna(),
                        games_actual.query('platform == "PC"')['user_score'].dropna(),
                         equal_var = False)

print('p-значение: ', pvalue)

if pvalue < alpha:
    display("Отвергаем нулевую гипотезу")
else:
    display("Не получилось отвергнуть нулевую гипотезу") 
```

Посчитаем средние пользовательские оценки их дисперсию и стандартное отклонение для платформ.

```python
pd.pivot_table(games_actual, values=['user_score'], index='platform', aggfunc=['mean', 'var', 'std'])
```

Полученное значение p-value близко к нулю, значит вероятность получить такое или большее различие случайно составляет менее 1%, т.е. почти никогда, поэтому мы отвергаем нулевую гипотезу.

Делаем вывод, что средние пользовательские рейтинги платформ Xbox One и PC не равны.

### Сравнение средних пользовательских рейтингов жанров

Сформулируем нулевую гипотезу: **'Средние пользовательские рейтинги жанров Action и Sports равны'**.

Сформулируем альтернативную гипотезу: **'Средние пользовательские рейтинги жанров Action и Sports не равны'**.

Оценим размеры выборок, после удаления пропусков.

```python
# определяем размер выборок для жанра Action
games_actual.query('genre == "Action"')['user_score'].dropna().shape[0]
```

```python
# определяем размер выборок для жанра Sports
games_actual.query('genre == "Sports"')['user_score'].dropna().shape[0]
```

Выборки не равны, значит укажем параметр equal_var = False. Применим t-test, для проверки нулевой гипотезы.

```python
alpha =.05   # критический уровень статистической значимости
#если p-value окажется меньше него - отвергнем гипотезу

tstat, pvalue = st.ttest_ind(games_actual.query('genre == "Action"')['user_score'].dropna(),
                        games_actual.query('genre == "Sports"')['user_score'].dropna(),
                         equal_var = False)

print('p-значение: ', pvalue)

if pvalue < alpha:
    display("Отвергаем нулевую гипотезу")
else:
    display("Не получилось отвергнуть нулевую гипотезу") 
```

Посчитаем средние пользовательские оценки их дисперсию и стандартное отклонение для жанров.

```python
pd.pivot_table(games_actual, values=['user_score'], index='genre', aggfunc=['mean', 'var', 'std'])
```
