## Предобработка данных

###  Обработка пропусков

Посчитаем, сколько в таблице пропущенных значений.

```python
# подсчет пропусков
realty.isna().sum()
```

Большего всего пропусков в столбце апартаменты. Как известно, апартаменты это таже квартира, только в ней нельзя прописаться. Поскольку этот столбец не относится к важным параметрам, для определения зависимости цены квартиры. То заменим все пропуски в столбце на 0.

```python
# заменяем пропуски в столбце на 0
realty['is_apartment'] = realty['is_apartment'].astype('bool')
realty['is_apartment'] = realty['is_apartment'].fillna(0)
```

Выделим основные столбцы, по которым будет происходить исследование, и в которых нужно обработать пропуски. Это столбцы: жилая площадь, площадь кухни, всего этажей в доме, расстояние до центра города и высота потолков.

Заменим пропуски в столбце жилая площадь медианными значениями сгруппированными по числу комнат.

```python
# заменяем пропуски в столбце жилая площадь с помощью метода transform()
realty['living_area'] = realty['living_area'].fillna(realty.groupby('rooms')['living_area'].transform('median'))
```

Теперь займемся пропусками в площади кухни. Тоже применим группировку с медианами по жилой площади.

```python
# заменяем пропуски в столбце площадь кухни с помощью метода transform()
realty['kitchen_area'] = realty['kitchen_area'].fillna(realty.groupby('living_area')['kitchen_area'].transform('median'))
```

Заменим пропуски в столбце всего этажей в доме на мединые по этажу

```python
# заменяем пропуски в столбце всего этажей с помощью метода transform()
realty['floors_total'] = realty['floors_total'].fillna(realty.groupby('floor')['floors_total'].transform('median'))
```

Проверим выполняется ли условие, что этаж квартиры больше числа этажей в доме.

```python
# в переменную floor_number сохраним условие (realty['floor'] > realty['floors_total'])
floor_number = realty.loc[(realty['floor'] > realty['floors_total'])]
display(floor_number)
```

Условие не выполняется, значит продолжаем исследование.

В столбце балконы 11519 пропусков. Скорее всего балконов нет и заменим пропуски на 0

```python
# заменяем значения на 0
realty['balcony'] = realty['balcony'].fillna(0)
```

Посмотрим на уникальные значения стобца `locality_name`

```python
# смотрим уникальные значения
realty['locality_name'].unique()
```

Оставим только названия населённых пунктов.

```python
# создадим список слов для удаления и оставим в столбце 'locality_name' только названия
remove_words = ['поселок', 'городской', 'село', 'тип', 'деревня', 'посёлок']
pat = '|'.join(remove_words)
realty['locality_name'] = realty['locality_name'].replace(pat,'', regex=True)
```

Заменим пропуски в столбце `city_centers_nearest` на мединые по названию населённого пункта

```python
# заполняем пропуски методом transform()
realty['city_centers_nearest'] = realty['city_centers_nearest'].fillna(realty.groupby('locality_name')['city_centers_nearest'].transform('median'))
```

Заменим пропуски в столбце `ceiling_height` на мединые по общему числу этажей

```python
# заменяем пропуски методом transform()
realty['ceiling_height'] = realty['ceiling_height'].fillna(realty.groupby('floors_total')['ceiling_height'].transform('median'))
```

Заменим пропуски в столбце 'days_expositiont' на мединые по общему цене квартиры

```python
# заменяем пропуски методом transform()
realty['days_exposition'] = realty['days_exposition'].fillna(realty.groupby('last_price')['days_exposition'].transform('median'))
```

Для дальнейшего удобства расчётов, переведем расстояния из метров в километры

```python
# переводим расстояния в километры
realty['city_centers_nearest'] = realty['city_centers_nearest'] / 1000 
```

Снова посчитаем пропуски

```python
# считаем пропуски
realty.isna().sum()
```

Не все пропуски удалось заменить.

Проверим данные на наличие дубликатов.

```python
# считаем дубликаты
realty.duplicated().sum()
```

###  Замена типа данных

Изменим типы данных в столбцах: цена, общая площадь, всего этажей, жилая площадь, площадь кухни, расстоние до центра города и сколько дней просуществовало объявление. С вещественных на целочисленные. Так же изменим тип данных в столбце дата подачи объявления, из строки сделаем дату.

## Расчёты и добавление результатов в таблицу

Расчитаем и добавим в таблицу столбец со стоимостью одного квадратного метра

День недели, месяц и год публикации объявления

```python
# добавим в таблицу 3 столбца: день, месяц, год побликации объявления
# используем функцию exec()
# выведем на экран первые 10 строк таблицы
for col in ['weekday', 'month', 'year']:
     exec('realty["' + col + '"] = realty["first_day_exposition"].dt.' + col)

realty.head(10)        
```

Добавим столбец с описанием этажа. Напишем функцию def floor_apartament (row), которая будет сравнивать значения из двух столбцов: этаж квартиры и всего этажей в доме. Если этаж первый - то она вернёт значение 'первый', если этаж совпадает с числом этажей в доме, тогда вернёт значение 'последний' и вернёт значение 'другой' во всех остальных случаях.

```python
# пишем функцию, которая будет возвращать описание этажа в зависимости от 2-х параметров,
# этажа квартиры и всего этажей в доме
# добавим получившиеся описание в новый столбец
def floor_apartament (row):
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
realty['floor_apartament'] = realty.apply(floor_apartament, axis=1) 
```

Добавим соотношение жилой площади к общей площади и площади кухни к общей площади.

```python
# считаем и добавляем соотношение жилой площади к общей площади
# считаем и добавляем соотношение площади кухни  к общей площади
realty['living_rate'] = realty['living_area'] / realty['total_area']
realty['kitchen_rate'] = realty['kitchen_area'] / realty['total_area']
```

###  Исследование факторов, влияющих на стоимость квартиры

Изучим, зависит ли цена от площади, числа комнат, удалённости от центра. Изучим зависимость цены от того, на каком этаже расположена квартира: первом, последнем или другом. Также изучим зависимость от даты размещения: дня недели, месяца и года.

Для изучения зависимости цены от площади, числа комнат, удалённости от центра, создадим сводную таблицу apartment_data. Посчитаем корреляцию цены квартиры от каждого из параметров методом corr(). И построим матрицу диаграмм рассеяния.

```python
# создаем сводную таблицу
# находим значение корреляции
apartment_data = realty.pivot_table(index=['price_for_m2'], 
                                    values=['last_price', 'total_area', 'rooms', 'city_centers_nearest'], aggfunc='mean')
apartment_data = apartment_data[['last_price', 'total_area', 'rooms', 'city_centers_nearest']]
apartment_data.corr().style.background_gradient(subset=['city_centers_nearest'], 
                                                cmap='Reds_r').background_gradient(subset=['last_price'],
                                                cmap='Blues_r').background_gradient(subset=['total_area'], 
                                                cmap='Oranges_r').background_gradient(subset=['rooms'],cmap='Greens')
```

Цена на прямую зависит от площади квартиры. Зависимость от числа комнат слабая. Так же слабая, но отрицательная зависимость от удаленности от центра.

Для изучения зависимости цены от даты размещения: дня недели, месяца и года, создадим сводную таблицу apartment_data_2. Посчитаем корреляцию цены квартиры от каждого из параметров методом corr(). И построим матрицу диаграмм рассеяния.

```python
# создаём сводную таблицу
# находим значение корреляции
apartment_data_2 = realty.pivot_table(index=['price_for_m2'], values=['last_price', 'weekday', 'month', 'year'], aggfunc='mean')
apartment_data_2 = apartment_data_2[['last_price', 'weekday', 'month', 'year']]
apartment_data_2.corr().style.background_gradient(subset=['last_price'], 
                                                cmap='Reds_r').background_gradient(subset=['weekday'],
                                                cmap='Blues_r').background_gradient(subset=['month'], 
                                                cmap='Oranges_r').background_gradient(subset=['year'],cmap='Greens')
```

```python
# строим матрицу диаграмм рассеяния
pd.plotting.scatter_matrix(apartment_data_2, figsize=(12, 12))
plt.show()
```

Цена не зависит от времени размещения. Кореляция везде почти равна 0.

Изучим зависимость цены от расположения квартиры на этаже. Для этого добавим столбец с числовым кодом, где 1- первый этаж, 2 - последний, 3 - другой этаж. Напишем для этого преобразования функцию floor_rage

```python
# пишем функцию floor_rage, которая на вход получает значение столбца 'floor_apartament' и возвращает значение:
# 1 если значение в столбце 'floor_apartament' = первый
# 2 если значение в столбце 'floor_apartament' = последний
# 3 если значение в столбце 'floor_apartament' = другой
def floor_rage (row):
    floor_apartament = row['floor_apartament']
    if floor_apartament == 'первый':
        return 1
    elif floor_apartament == 'последний':
        return 2
    elif floor_apartament == 'другой':
        return 3
    else:
        return 4
realty['floor_rage'] = realty.apply(floor_rage, axis=1) 
```

```python
# создаём сводную таблицу
# находим значение корреляции
apartment_data_3 = realty.pivot_table(index=['price_for_m2'], values=['last_price', 'floor_rage'], aggfunc='mean')
apartment_data_3.corr()
```

Цена квартиры не зависит от этажа.

**Вывод** 
* цена квартиры сильнее всего зависит от площади квартиры 
* цена квартиры так же зависит, но не так сильно, от числа комнат и удалённости от центра
* цена квартиры не зависит от даты подачи объявления и от этажа квартиры
