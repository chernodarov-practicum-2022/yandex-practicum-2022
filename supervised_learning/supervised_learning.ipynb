# Описание проекта

Из «Бета-Банка» стали уходить клиенты. Каждый месяц. Немного, но заметно. Банковские маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.
Нужно спрогнозировать, уйдёт клиент из банка в ближайшее время или нет. Вам предоставлены исторические данные о поведении клиентов и расторжении договоров с банком.

## Подготовка данных

Импортируем нужные библиотеки.

```python
# импортируем нужные библиотеки
import pandas as pd
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.utils import shuffle
import seaborn as sns
```

Прочитаем файл /datasets/Churn.csv и сохраним его в переменной `bank`

```python
# читаем файл и сохраняем его в bank
bank = pd.read_csv('/datasets/Churn.csv') 
```

Напишем функцию, которая выводит общую информацию о таблице, а так же посчитает пропуски и проверит на наличие дубликатов.

```python
#пишем функцию summary
def summary(df):
    print(df.info())
    display(df.head())
    count=0
    for col in df.columns:
        if df.isna().sum()[col]==0:
            count+=1
        else:
            print(f'Пропущенных значений в столбце {col}: {df.isna().sum()[col]}')
    if count==len(df.columns):
        print('Пропущенных значений в таблице нет')
    print(f'Повторов в таблице {df.duplicated().sum()}')  
    
summary(bank) 
```

Итак в таблице 14 столбцов и 10000 строк о клиентах «Бета-Банка». Согласно документации к данным:

* RowNumber — индекс строки в данных
* CustomerId — уникальный идентификатор клиента
* Surname — фамилия
* CreditScore — кредитный рейтинг
* Geography — страна проживания
* Gender — пол
* Age — возраст
* Tenure — сколько лет человек является клиентом банка
* Balance — баланс на счёте
* NumOfProducts — количество продуктов банка, используемых клиентом
* HasCrCard — наличие кредитной карты
* IsActiveMember — активность клиента
* EstimatedSalary — предполагаемая зарплата
* Exited — факт ухода клиента

Столбцы `RowNumber`, `CustomerId`, `Surname`, никак не могу повлиять на решение клиента об уходе из банка. Удалим их.

```python
# удаляем столбцы 'RowNumber', 'CustomerId', 'Surname'
bank.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

Следуя правилам хорошего тона, приведём названия столбцов к нижнему регистру.

```python
# производим замены в названии столбцов
bank.columns = bank.columns.str.lower()
```

В столбце `tenure` 909 пропусков. Это почти 10% от всего датафрейма, удалять их нельзя. Заменим медианными значениями, сгруппированными по столбцу `numofproducts`.

```python
# производим замену пропусков
bank['tenure'] = bank['tenure'].fillna(bank.groupby('numofproducts')['tenure'].transform('median'))
```

Столбцы `gender` и `geography` преобразуем в численные признаки методом прямого кодирования, или отображения (англ. One-Hot Encoding, OHE) с помощью функции pd.get_dummies(). Укажем аргумент drop_first, чтобы удалить первый столбец при выполнении кодировки. 

Посмотрим на уникальные значения этих столбцов.

Разделим исходные данные на обучающую, валидационную и тестовую выборки. Из соотношения:

```python
# масштабируем признаки
# в переменную numeric сохраним признаки ,которые нужно отмасштабировать
# преобразуем обучающую, влидационную и тестовую выборки функцией transform()
numeric = ['creditscore', 'age', 'tenure', 'balance', 'numofproducts', 'estimatedsalary']

scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

### Дерево решений

Возьмём модель дерево решений и определим для неё гиперпараметры.

```python
# берём модель дерево решений, обучаем её и находим для неё гиперпараметры с помощью GridSearchCV
model = DecisionTreeClassifier(random_state=12345)
parametrs = { 'max_depth': range (1,21, 2),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2) }
grid = GridSearchCV(model, parametrs)
grid.fit(features_train, target_train)
grid.best_params_
```

Найдём полноту для модели решающего дерева с полученными гиперпараметрами. Полнота выявляет, какую долю положительных среди всех ответов выделила модель

```python
# находим полноту модели решающего дерева
model_decision_tree = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree.fit(features_train, target_train)
predicted_valid_decision_tree = model_decision_tree.predict(features_valid)
recall_score_model_decision_tree = recall_score(target_valid, predicted_valid_decision_tree)
display('Полнота модели решающего дерева:', recall_score_model_decision_tree)
```

Найдём точность модели решающего дерева с полученными гиперпараметрами. Точность определяет, как много отрицательных ответов нашла модель, пока искала положительные. Чем больше отрицательных, тем ниже точность.

```python
# находим точность модели решающего дерева
model_decision_tree = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree.fit(features_train, target_train)
predicted_valid_decision_tree = model_decision_tree.predict(features_valid)
precision_score_model_decision_tree = precision_score(target_valid, predicted_valid_decision_tree)
display('Точность модели решающего дерева:', precision_score_model_decision_tree)
```

Найдём F1-меру (это среднее гармоническое полноты и точности) модели решающего дерева с полученными гиперпараметрами.

```python
# находим F-1 меру модели решающего дерева
model_decision_tree = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree.fit(features_train, target_train)
predicted_valid_decision_tree = model_decision_tree.predict(features_valid)
f1_score_model_decision_tree = f1_score(target_valid, predicted_valid_decision_tree)
display('F-1 модели решающего дерева:', f1_score_model_decision_tree)
```

```python
# найдём значение AUC-ROC модели решающего дерева
model_decision_tree = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree.fit(features_train, target_train)
probabilities_valid_decision_tree = model_decision_tree.predict_proba(features_valid)
probabilities_one_valid_decision_tree = probabilities_valid_decision_tree[:, 1]

model_decision_tree = DecisionTreeClassifier(random_state=12345, max_depth=5,
                                            min_samples_leaf=6, min_samples_split=2).fit(features_train, target_train)
predicted_valid_decision_tree = model_decision_tree.predict(features_valid)
auc_roc = roc_auc_score(target_valid, probabilities_one_valid_decision_tree)
display(auc_roc)
```

### Случайный лес

Возьмём модель случайный лес и определим для неё гиперпараметры.

```python
# берём модель случайный лес, обучаем её и находим для неё гиперпараметры с помощью GridSearchCV
model = RandomForestClassifier(random_state=12345)
parametrs = { 'n_estimators': range (1,51, 2),
              'max_depth': range (1,15, 2),
              'min_samples_split': range (2,10,2) }
grid = GridSearchCV(model, parametrs)
grid.fit(features_train, target_train)
grid.best_params_
```

Найдём полноту для модели случайного леса с полученными гиперпараметрами.

```python
# находим полноту модели случайного леса
model_random_forest = RandomForestClassifier(random_state=12345, **grid.best_params_)
model_random_forest.fit(features_train, target_train)
predicted_valid_random_forest = model_random_forest.predict(features_valid)
recall_score_model_random_forest = recall_score(target_valid, predicted_valid_random_forest)
display('Полнота модели случайный лес:', recall_score_model_random_forest)
```

Найдём точность модели случайного леса с полученными гиперпараметрами.

```python
# находим точность модели случайного леса
model_random_forest = RandomForestClassifier(random_state=12345, **grid.best_params_)
model_random_forest.fit(features_train, target_train)
predicted_valid_random_forest = model_random_forest.predict(features_valid)
precision_score_model_random_forest = precision_score(target_valid, predicted_valid_random_forest)
display('Точность модели случайного леса:', precision_score_model_random_forest)
```

Найдём F1-меру (это среднее гармоническое полноты и точности) модели случайного леса с полученными гиперпараметрами.

```python
# находим F-1 меру модели случайного леса
model_random_forest  = RandomForestClassifier(random_state=12345, **grid.best_params_)
model_random_forest.fit(features_train, target_train)
predicted_valid_random_forest = model_random_forest.predict(features_valid)
f1_score_model_random_forest = f1_score(target_valid, predicted_valid_random_forest)
display('F-1 модели случайного леса:', f1_score_model_random_forest)
```

```python
# найдём значение AUC-ROC модели случайного леса
model_random_forest = RandomForestClassifier(random_state=12345,**grid.best_params_)
model_random_forest.fit(features_train, target_train)
probabilities_valid_random_forest = model_random_forest.predict_proba(features_valid)
probabilities_one_valid_random_forest = probabilities_valid_random_forest[:, 1]

model_random_forest = RandomForestClassifier(random_state=12345, max_depth=13,
                                            n_estimators=49, min_samples_split=8).fit(features_train, target_train)
predicted_valid_random_forest = model_random_forest.predict(features_valid)
auc_roc = roc_auc_score(target_valid, probabilities_one_valid_random_forest)
display(auc_roc)
```

### Логистическая регрессия

Найдём полноту для модели логистической регрессии.

## Борьба с дисбалансом

С дисбалансом классов будем бороться 3 путями: взвешивание классов, увеличение выборки, уменьшение выборки.

### Дерево решений

#### Взвешивание классов

```python
# берём модель дерево решений, обучаем её и находим для неё гиперпараметры с помощью GridSearchCV
model = DecisionTreeClassifier(random_state=12345, class_weight='balanced')
parametrs = { 'max_depth': range (1,21, 2),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2) }
grid = GridSearchCV(model, parametrs)
grid.fit(features_train, target_train)
grid.best_params_
```

```python
# находим F-1 меру модели решающего дерева при взвешивании классов
model_decision_tree_weight = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree_weight.fit(features_train, target_train)
predicted_valid_decision_tree_weight = model_decision_tree_weight.predict(features_valid)
f1_score_model_decision_tree_weight = f1_score(target_valid, predicted_valid_decision_tree_weight)
display('F-1 модели решающего дерева:', f1_score_model_decision_tree_weight)
```

F-1 мера без взвешивания классов у модели Дерева решения была: 0.6104651162790699

#### Увеличение выборки

```python
# увеличим выборку
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
```

```python
best_model = None
best_result = 0
best_depth = 0
best_leaf = 0
best_split = 0
for leaf in range(1,8):
    for depth in range(1, 21, 2):
        for split in range(2, 10,2):
            model_decision_tree_up = DecisionTreeClassifier(random_state=12345, min_samples_leaf=leaf, max_depth=depth,
                                                            min_samples_split=split)
            model_decision_tree_up.fit(features_upsampled, target_upsampled)
            predicted_valid_decision_tree_up = model_decision_tree_up.predict(features_valid)
            f1_score_model_decision_tree_up = f1_score(target_valid, predicted_valid_decision_tree_up)
            if f1_score_model_decision_tree_up > best_result:
                best_model = model_decision_tree_up
                best_result = f1_score_model_decision_tree_up
                best_depth = depth
                best_leaf = leaf
                best_split = split
                
display('F-1 модели случайного леса:', best_result)    
```

Метрика F-1 тоже стала меньше.

#### Уменьшение выборки

```python
# уменьшим выборку
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.1)
```

```python
# берём модель дерево решений, обучаем её и находим для неё гиперпараметры с помощью GridSearchCV
model = DecisionTreeClassifier(random_state=12345)
parametrs = { 'max_depth': range (1,21, 2),
              'min_samples_leaf': range (1,8),
              'min_samples_split': range (2,10,2) }
grid = GridSearchCV(model, parametrs)
grid.fit(features_downsampled, target_downsampled)
grid.best_params_
```

```python
model_decision_tree_down = DecisionTreeClassifier(random_state=12345, **grid.best_params_)
model_decision_tree_down.fit(features_downsampled, target_downsampled)
predicted_valid_decision_tree_down = model_decision_tree_down.predict(features_valid)
f1_score_model_decision_tree_down = f1_score(target_valid, predicted_valid_decision_tree_down)
display('F-1 модели решающего дерева:', f1_score_model_decision_tree_down)
```

Метрика F-1 тоже стала меньше.

#### Изменение порога
